```julia
using Pkg
Pkg.add("PyCall")
using PyCall
np=pyimport("numpy")
```

```julia
x = Matrix(data[:,[2,3,4,5]])
```

```julia
y = Matrix(data[:,[6]])
```

```julia
using ScikitLearn
@sk_import linear_model: LogisticRegression
```

```julia
log_reg_model = LogisticRegression()
```

```julia
fit!(log_reg_model,x,y)
```

```julia
predictions = predict(log_reg_model,x)
```

```julia
# Checking Accuracy
@sk_import metrics:accuracy_score
```

```julia
accur = accuracy_score(predictions,y)
```

```julia
# Decision Tree Classifier
@sk_import tree:DecisionTreeClassifier

tree_model = DecisionTreeClassifier()
```

```julia
fit!(tree_model,x,y)
```

```julia
tree_predictions = predict(tree_model,x)
```

```julia
tree_accuracy = accuracy_score(tree_predictions,y)
```

```julia
#Random Forest Classification
@sk_import ensemble:RandomForestClassifier
rf_model = RandomForestClassifier()
fit!(rf_model,x,y)
rf_predictions=predict(rf_model,x)
rf_accuracy=accuracy_score(rf_predictions,y)
```

```julia
# Split into train and test set
@sk_import model_selection:train_test_split
```

```julia
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```julia
log_reg_2 = LogisticRegression()
```

```julia
fit!(log_reg_2,x_train,y_train)
```

```julia
prediction_log_reg=predict(log_reg_2,x_test)
```

```julia
accuracy_log_reg=accuracy_score(prediction_log_reg,y_test)
```
